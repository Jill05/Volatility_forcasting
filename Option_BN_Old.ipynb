```python
option_data['strike_price']=  option_data['Ticker'].apply(lambda x: ''.join(filter(str.isdigit, x)))

print(option_data)
```

```python
# Create Expiry column
option_data['Expiry'] = option_data['Date'] + pd.offsets.Week(weekday=4)

# Days difference
option_data['Days_to_Expiry'] = (
    option_data['Expiry'] - option_data['Date']
).dt.days

print(option_data.tail())
```

```python
daily_option = option_data.groupby(['Date','Ticker']).agg({
        'Open': 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        'Call/Put': 'first',
        'strike_price' : 'first',
        'Expiry' : 'first',
        'Days_to_Expiry' : 'first' 
})
```

```python
def calculate_iv(row):

    S = row['Open_y']              # underlying price
    K = row['strike_price']        # strike
    r = 0.065                      # interest rate %
    q = 0                          # dividend yield
    dte = row['Days_to_Expiry']    # days to expiry
    option_price = row['Open_x']
    try:
        if row['Call/Put'] == 'CE':
            me = BS.Me(
                [S, K, r, q, dte],
                callPrice=option_price)
        else:
            me = BS.Me(
                [S, K, r, q, dte],
                putPrice=option_price)
        return me.impliedVolatility
    except:
        return None
# apply row-wise
main_dataset['IV'] = main_dataset.apply(calculate_iv, axis=1)

print(main_dataset)
```
